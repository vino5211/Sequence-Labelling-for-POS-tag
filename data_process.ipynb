```python
import re

def convert_to_bio(text):
    text = text.replace(' ','')
    tokens_with_tags = text.split('/')
    # print(tokens_with_tags)
    words = []
    tags = []
    for i in range(0, len(tokens_with_tags) - 1, 2):
        word = tokens_with_tags[i]
        tag = tokens_with_tags[i+1]  
        if not re.match(r'^[A-Za-z]+$', tag):
            print(tag,'\t',text)
            
        for j, char in enumerate(word):
            if j == 0:
                words.append(char)
                tags.append(f'B-{tag}')
            else:
                words.append(char)
                tags.append(f'I-{tag}')

    # Zip the words and tags to generate CoNLL format
    conll_format = list(zip(words, tags))
    return conll_format

input_text = "豫 讓 /nr/逃 /v/到 /v/山 中 /n/， /w/説 /v/： /w/“ /w/唉 /y/！ /w/士 /n/爲 /p/知 /v/己 r/者 /r/死 /v/， /w/女 /n/爲 /p/悦 /v/己 /r/者 /r/容 /n/。 /w"
bio_data = convert_to_bio(input_text)
print(bio_data)
```

```python
def get_distinct_entry(text):
    text = text.replace(' ','')
    tokens_with_tags = text.split('/')
    # print(tokens_with_tags)
    entrys = set()
    for i in range(0, len(tokens_with_tags) - 1, 2):
        word = tokens_with_tags[i]
        tag = tokens_with_tags[i+1]  
        entrys.add((word, tag))
    return entrys
```

### 抽取地名

```python
import pandas as pd
df2 = pd.read_excel('all_round2.xlsx')
def get_location(text):
    if type(text) != str:
        return None
    text = text.replace(' ','')
    tokens_with_tags = text.split('/')
    # print(tokens_with_tags)
    loc = []
    for i in range(0, len(tokens_with_tags) - 1, 2):
        word = tokens_with_tags[i]
        tag = tokens_with_tags[i+1]  
        if tag == 'ns':
            loc.append(word)

    if len(loc) == 0:
        return None
    return loc
df2['classical_location'] = df2['原文'].apply(get_location)
df2['modern_location'] = df2['译文'].apply(get_location)
df2.to_excel('all_round2_with_location.xlsx', index=False)
```

```python
from collections import Counter
dynasty_location_counts  = {}
# Filter rows where 'classical_location' is not None and '来源' contains any of the specified keywords
keywords = ['唐书', '宋史', '元', '明']
for kw in keywords:
   
    filtered = df2[df2['classical_location'].notna() & df2['来源'].str.contains(kw)]

    # Flatten the list of locations and count occurrences
    locs = [loc for sublist in filtered['classical_location'] if isinstance(sublist, list) for loc in sublist]
    location_counts = Counter(locs)
    dynasty_location_counts[kw] = location_counts
print(location_counts)
```

```python
import pandas as pd

rows = []
for dynasty, counter in dynasty_location_counts.items():
    for location, count in counter.items():
        rows.append({'dynasty': dynasty, 'location': location, 'counts': count})

df_counts = pd.DataFrame(rows)
df_counts.to_excel('dynasty_location_counts.xlsx', index=False)
```

### 抽取长度超过10的词

```python
import pandas as pd
df2 = pd.read_excel('all_round2.xlsx')
def get_words_longer_than_10(text):
    if type(text) != str:
        return None
    text = text.replace(' ','')
    tokens_with_tags = text.split('/')
    # print(tokens_with_tags)
    words = []
    for i in range(0, len(tokens_with_tags) - 1, 2):
        word = tokens_with_tags[i]
        tag = tokens_with_tags[i+1]  
        if len(word) >= 10:
            words.append((word,tag))

    if len(words) == 0:
        return None
    return words
df2['classical_words'] = df2['原文'].apply(get_words_longer_than_10)
df2['modern_words'] = df2['译文'].apply(get_words_longer_than_10)
df2.to_excel('all_round2_words_longer_than_10.xlsx', index=False)
```

```python
filtered = df2[df2['modern_words'].notna() ]

# Flatten the list of locations and count occurrences
words = [loc for sublist in filtered['modern_words'] if isinstance(sublist, list) for loc in sublist]
counts = Counter(words)


rows = []

for word, count in counts.items():
    rows.append({ 'word': word[0],'tag':word[1], 'counts': count})

df_counts = pd.DataFrame(rows)
df_counts.to_excel('words_longer_than_10.xlsx', index=False)
```
